```python
import numpy as np
import matplotlib.pyplot as plt

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = np.copy(x0)
    convergence = []

    for k in range(max_iterations):
        x_old = np.copy(x)

        for i in range(n):
            sum1 = sum(A[i][j] * x[j] for j in range(i))
            sum2 = sum(A[i][j] * x_old[j] for j in range(i + 1, n))
            x[i] = (b[i] - sum1 - sum2) / A[i][i]

        # Calcular e armazenar a norma da diferena
        norm_diff = np.linalg.norm(x - x_old, ord=np.inf)
        convergence.append(norm_diff)

        # Verificar a convergncia
        if norm_diff < tol:
            break

    return x, convergence

# Parmetros do problema
R1, R2, R3, R4, R5, R6 = 4, 1, 3, 2, 5, 1
V1, V2 = 8, 5
A = np.array([[R1 + R2, -R2, 0], [-R2, R2 + R3 + R4, -R4], [0, -R4, R4 + R5 + R6]], dtype=float)
b = np.array([V1, 0, V2], dtype=float)
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100

# Resolver o sistema e obter os dados de convergncia
x, convergence = gauss_seidel(A, b, x0, tol, max_iterations)

# Plotar o grfico de convergncia
plt.plot(convergence, marker='o', linestyle='-')
plt.yscale('log')
plt.xlabel('Iteraes')
plt.ylabel('Norma da Diferena')
plt.title('Convergncia do Mtodo de Gauss-Seidel')
plt.grid(True)
plt.show()
```
