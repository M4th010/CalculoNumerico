```python
from math import pi, exp
from numpy import array, divide, linalg
Icr = 31.9824*10**(-9)
n = 2
k = 1.3806*10**(-23)
T = 300
q = 1.6022*10**(19)
V = 24
R = 10

x = array([[1],[2]])
iter = 0
maxit = 50
es = 0.001
while True:
  f1 = (Icr((exp**(x(2)/(n*k*T/q)-1))))-x(1)
  f2 = ((V-x(2))/R)-x(1)
  F = array([f1,f2])       # Matriz de funes

  J = array([[-1, Icr*(exp**(x(2)/(n*k*T/q))-1)/(n*k*T/q)],
             [-1,                                -1/R]])  # Matriz Jacobiana
  dx = linalg.lstsq(J,F,rcond=None)[0]
  x = x - dx
  iter += 1
  ea = max(abs(divide(dx,x)))
  if iter >= maxit or ea <= es:
    break
print(f'Com {iter} iteraes: \n x = \n {x}')
```

```python
from math import pi, exp
from numpy import array, divide, linalg
lf = 0.4
le = 0.0008
N = 200
I = 10
mi = 4 * pi * 10**(-7)
x = array([[1250000],
           [70]])
iter = 0
maxit = 50
es = 0.001
while True:
  f1 = (2*x[0]*le + x[1]*lf - N*I)
  f2 = (-mi*x[0] + 1.8*(1-exp(-(x[1])/40)))
  F = array([f1,f2])

  J = array([[2*le,                        lf],
              [-mi, 1.8*(1/40)*exp(-(x[1])/40)]])
  dx = linalg.lstsq(J,F,rcond=None)[0]
  x = x - dx
  iter += 1
  ea = max(abs(divide(dx,x)))
  if iter >= maxit or ea <= es:
    break
print(f'Com {iter} iteraes: \n x = \n {x}')
```
