```
from math import pi, exp
from numpy import array, divide, linalg

x = array([0.3,0.3])
iter = 0
maxit = 50
es = 0.0001
while True:
  f1 = (x[0]**4+0.06823*x[0])-(x[1]**4+0.05848*x[1])-0.01509
  f2 = (x[0]**4+0.05848*x[0])-(x[1]**4+0.11696*x[1])
  F = array([f1,f2])

  J = array([[4*x[0]**3+0.06823,  -4*x[1]**3-0.05848],
              [4*x[0]**3+0.05848,           -8*x[1]**3-0.11696]])
  dx = linalg.lstsq(J,F,rcond=None)[0]
  x = x - dx
  iter += 1
  ea = max(abs(divide(dx,x)))
  if iter >= maxit or ea <= es:
    break
print(f'Com {iter} iteraes: \n x = \n {x}')
```
