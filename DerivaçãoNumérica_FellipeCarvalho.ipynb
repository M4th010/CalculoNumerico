```python
# Derivao numrica


#Importa a biblioteca necessria
from numpy import divide, diff
import matplotlib.pyplot as plt

#Parmetros da questo
t = [0, 0.5, 1, 2, 3]
V = [100, 50, 35, 9.5, 0.05]
d = divide(diff(V), diff(t))
C = -0.01  
I = C * d
tt = [0.25, 0.75, 1.5, 2.5]

#Imprime na tela resultado
print(f'd = {d}\n')
print(f'V = {V}\n')

#Visualizao plotagem do grfico
fig, ax = plt.subplots()
Corr = ax.plot(t, V, linewidth=1, marker='s', label='Tenso medida', linestyle='--', color='k')
ax2 = ax.twinx()  # cria nova escala no lado oposto do grfico eixo y
Tens = ax2.plot(tt, I, linewidth=1, marker='d', label='Corrente estimada')
lns = Tens + Corr
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Tenso (V)')
ax2.set_ylabel('Corrente (A)')
ax2.yaxis.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Derivao numrica
# Gradiente

#importa a biblioteca 
import matplotlib.pyplot as plt
import numpy as np


#resoluo da imagem
x_range = np.arange(-2, 2, 0.2)  # intervalo e quantidade de setas
y_range = np.arange(-2, 2, 0.2)  # intervalo e quantidade de setas
x_mesh, y_mesh = np.meshgrid(x_range, y_range)
# linhas de campo
E = np.multiply(-x_mesh, np.exp(-(np.power(x_mesh, 2)) - np.power(y_mesh, 2)))
dy, dx = np.gradient(-E)
plt.figure(figsize=(10, 10))  # Tamanho
plt.quiver(x_mesh, y_mesh, -dx, -dy, pivot="middle")  # Setas
plt.axis("scaled")  # Ajuste automtico de escala
# superfcies equipotenciais geradas pelas duas cargas
xp = np.arange(-2, 2, 0.1)
yp = np.arange(-2, 2, 0.1)
x_mesh, y_mesh = np.meshgrid(xp, yp)
E = np.multiply(x_mesh, np.exp(-(np.power(x_mesh, 2)) - np.power(y_mesh, 2)))
d_o = 4
# lao de repetio
if (len(E) % 2) != 0:
    l = int((len(E) + 1) / 2)
else:
    l = int((len(E)) / 2)
niveis = np.linspace(E[l][l - d_o], -E[l][l - d_o], 6)  # niveis intersecces do plano
eqp = plt.contour(x_mesh, y_mesh, -E, levels=niveis)  # linhas de contorno
plt.clabel(eqp)
# posicionamento das duas cargas
# carga eltrica negativa
carga_x1 = 0.70
carga_y1 = 0
# carga eltrica positiva
carga_x2 = -0.70
carga_y2 = 0
plt.plot(carga_x1, carga_y1, color='blue', marker='o', linestyle='dashed', markersize=7)
plt.plot(carga_x2, carga_y2, color='red', marker='o', linestyle='dashed', markersize=7)
plt.title('Campo Eltrico')
# plt.savefig("Campo_eltrico.png")
plt.show()
```
