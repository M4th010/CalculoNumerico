```python
from math import cosh

def f(
    C,
    d = 500
):
  return C*(cosh(d/(2*C))-1)-50
def bisseco(
    start,
    end,
    tol,
    iter_max = 16
):
  half_point = (start + end) / 2
  if f(start) * f(end) > 0:
    print("Nenhuma raz encontrada.")
  else:
    c = 0
    Erro = abs(f(end)-f(start))
    while Erro > tol and c < iter_max:
      half_point = (start + end ) / 2.0
      if f(half_point) == 0:
        return [half_point, c]
      elif f(start) * f(half_point) < 0:
        end = half_point
        c += 1
      else:
        start = half_point
      Erro = abs(f(end)-f(start))
    return {"half_point": half_point, "iterao": c}

resp = bisseco(start = 600, end=650, tol=0.0001, iter_max=50)
print(f'raz aprox { resp ["half_point"]:.4f}')
print(f'O nmero de iteraes foi {resp["iterao"]}')
```

```python
import math as m
import numpy as np
import sympy as sp

def f(
    B,
    f = 1000,
    L = 0.1,
    R = 1000
):
  tan_teta = (2*m.pi*f*L)/R
  teta = m.atan(tan_teta)
  return m.sin(B - teta)+m.sin(teta)*m.exp(-(B/tan_teta))

def d(
    B,
    f = 1000,
    L = 0.1,
    R = 1000
):
  tan_teta = (2*m.pi*f*L)/R
  teta = m.atan(tan_teta)
  return m.cos(B - teta)-m.exp(-B*sp.cot(teta))*m.sin(teta)*sp.cot(teta)

def newtonRaphson ( x , # x inicial
                   maxit =50, # nmero mximo de iteraes
                    TOL=0.0001) : # erro tolerado

  a = f(x)/d(x)
  iter = 0
  while abs(a) > TOL and iter < maxit : # critrio de parada
    print(abs(a))
    a = f(x)/d(x)
    # x(i+1) = x(i)f(x)/f'(x)
    x = x - a
    iter+= 1
  return[x,iter]

# Exemplo

x0 = 200 # Valor inicial
ans = newtonRaphson(x0)
raiz = ans [0]

print(f'A raiz  {raiz:.5f}. Com {ans[1]} iteraes.' )
```
